```python
import pandas as pd


def fetch_settings(fp: str):
    """Fetch settings from excel file"""
    sheet_names = ["data", "layout", "calc"]
    setting_df_dict = pd.read_excel(fp, sheet_name=sheet_names)
    data_df = setting_df_dict["data"]
    layout_df = setting_df_dict["layout"]
    calc_df = setting_df_dict["calc"]

    graph_df = pd.read_excel(fp, sheet_name="graph")

    return data_df, layout_df, graph_df, calc_df
```

```python
setting_file_path = r"C:\Users\tomon\Documents\Python\settings.xlsx"
data_df, layout_df, graph_df, calc_df = fetch_settings(fp=setting_file_path)
```
